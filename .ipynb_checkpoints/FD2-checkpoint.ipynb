### Step 4: Acquiring a Method of Calculating Dependencies
We'd also like to know if any of our features are significantly dependent on the values of other feature, implying redundancies.

For this step we use Theil's U method of calculating categorical asymmetrical dependencies, with code copied from this implementation: https://www.kaggle.com/code/shakedzy/alone-in-the-woods-using-theil-s-u-for-survival

The copied code includes the function conditional_entropy which is merely a stepping stone for the final function theil_u which we'll be using directly. The theil_u function takes 2 features as input, in the form of the lists, and outputs the rate at which y can predict x. As such, we have an asymmetric output, requiring that we run the theil_u function over all combinations of features and _in both directions_.

```python
# copied from above-mentioned source

def conditional_entropy(x,y):
    # entropy of x given y
    y_counter = Counter(y)
    xy_counter = Counter(list(zip(x,y)))
    total_occurrences = sum(y_counter.values())
    entropy = 0
    for xy in xy_counter.keys():
        p_xy = xy_counter[xy] / total_occurrences
        p_y = y_counter[xy[1]] / total_occurrences
        entropy += p_xy * math.log(p_y/p_xy)
    return entropy

def theil_u(x,y):
    s_xy = conditional_entropy(x,y)
    x_counter = Counter(x)
    total_occurrences = sum(x_counter.values())
    p_x = list(map(lambda n: n/total_occurrences, x_counter.values()))
    s_x = ss.entropy(p_x)
    if s_x == 0:
        return 1
    else:
        return (s_x - s_xy) / s_x
```

```python
# applying Theil's U

dimensions = len(ch1_df.columns)
dependencies = np.zeros((dimensions, dimensions))

for x_index, x in enumerate(ch1_df.columns):
    for y_index, y in enumerate(ch1_df.columns):
        result = theil_u(ch1_df[x].to_list(), ch1_df[y].to_list())
        dependencies[x_index][y_index] = result
        print(f"Prediction rate of {x} using {y} is {result:.2f}.")
        
```

### Step 6: Analysing Theil's U Dependency
Due to very many features, even the above illustration is a little much. Therefore, we try to query our results for 2 key possibilities:

1. Any dependencies between features greater than 0.67
2. Any predictability of mushroom edibility greater than 0.1, by any feature.

The first query is our boundary for what we feel is an alarming level of dependency. While the second query is chosen in case any features are particularly good at predicting edibility, as we would not want to wrongly manipulate these.

```python
d_len = len(dependencies)
for x in range(d_len):
    for y in range(d_len):
        x_col = ch1_df.columns[x]
        y_col = ch1_df.columns[y]
        d_val = dependencies[x][y]
        
        # we skip looking at mushroom edibility predicting other features' values
        if(y_col != "poisonous?" and y_col != x_col):
            if(d_val >= 0.67 or (d_val >= 0.1 and d_val == "poisonous?")):
                print(f"{y_col} predicts {x_col}: {d_val}")
```

### Step 11: Show_Training_Results and Basic RidgeClassifier

Since we'll be displaying the results of training-algorithms often throughout this file, we define a function to do this for us.

```python
def show_training_results(model_title, model, X_train, y_train, X_dev, y_dev):
    model_train_pred = model.predict(X_train)
    model_dev_pred = model.predict(X_dev)

    print(f"{model_title} - training set")
    print(f"Accuracy: {accuracy_score(y_train, model_train_pred):.3}")
    print(f"Confusion matrix:\n{confusion_matrix(y_train, model_train_pred)}")

    print("\n")

    print(f"{model_title} - dev set")
    print(f"Accuracy: {accuracy_score(y_dev, model_dev_pred):.3}")
    print(f"Confusion matrix:\n{confusion_matrix(y_dev, model_dev_pred)}")
```

### Step 12: Grid Search and Cross-validation over RidgeClassifier

The default training already provides results at or near 100%, leaving little room for improvements.

Still, now we'll try grid search with cross-validation. And since we'll be doing this repeatedly over the different algorithms, we decide to make a couple of utility functions both for training and displaying GridSearchCV.

```python
def train_gscv(estimator, params):
    gscv = GridSearchCV(
        estimator=estimator,
        param_grid=params,
        scoring="accuracy",
        n_jobs=-1,
        cv=10,
        return_train_score=True)
    gscv.fit(ch1_X_train, ch1_y_train)
    return gscv
```

```python
def show_gscv_results(model_title, gscv):
    print(f"Best params: {gscv.best_params_}")
    print(f"Best score: {gscv.best_score_:.3f}")
    mean_training_score = gscv.cv_results_['mean_train_score']
    print(f"Mean Training Score: {np.around(mean_training_score, decimals=3)}")
    mean_testing_score = gscv.cv_results_['mean_test_score']
    print(f"Mean Testing Score: {np.around(mean_testing_score, decimals=3)}")
    
    print("\n")

    show_training_results(
        model_title,
        gscv.best_estimator_,
        ch1_X_train, ch1_y_train,
        ch1_X_dev, ch1_y_dev)
```

```python
params = {
    "alpha": [0, 0.001, 0.01, 0.1, 1.0, 10.0],
    "solver": ["svd", "cholesky", "sparse_cg", "lsqr"],
    "random_state": [42]
}

ridge_gscv = train_gscv(RidgeClassifier(), params)
```

```python
show_gscv_results("Ridge/Grid Search", ridge_gscv)
```

### Step 23: Inspecting The Decision Trees

Before we run a grid search, we'd like to know a few things about the trees we've produced. In particular, how deep do they go? How many leaves do they have? We calculate a few such basic statistics below.

```python
num_estimators = 0

sum_leaves = 0
max_leaves = 0

sum_depth = 0
max_depth = 0

for estimator in rfc.estimators_:
    num_estimators += 1
    t = estimator.tree_
    sum_depth += t.max_depth
    if(max_depth < t.max_depth):
        max_depth = t.max_depth
    sum_leaves += t.n_leaves
    if(max_leaves < t.n_leaves):
        max_leaves = t.n_leaves
        
avg_depth = sum_depth / num_estimators
avg_leaves = sum_leaves / num_estimators

print(f"Num Estimators: {num_estimators}")
print(f"Max Depth: {max_depth}")
print(f"Avg Depth: {avg_depth}")
print(f"Max Leaves: {max_leaves}")
print(f"Avg Leaves: {avg_leaves}")
```

## Chapter 2: On The Sentiment of Norwegian Reviews (Task 2 of 3)

We want to train a ML model to accurately predict the sentiments ("Positive", "Negative" or "Neutral") for 10565 already labeled sentences.

```python
ch2_data = {}
for name in ["train", "dev", "test"]:
    file_path = "Data\\" + "3class\\" + f"{name}.json"
    with open(file_path) as filedata:
        ch2_data[name] = json.load(filedata)

print('Size of the 3 different sets: ')
print(len(ch2_data['train']))
print(len(ch2_data['dev']))
print(len(ch2_data['test']))
```

### Step 2: Preprocessing Plan

We need to make sure our data is consistent and in a format we can train on. Therefore, going forward we're going to have a range of new functions, which will try to:

1. Translate any english sentences to norwegian
2. Clean the data of i.e. stopwords, and reducing to lowercase
3. and lemmatize words

After that, we'll also bring our data to the right format for training. For this we'll need to:

1. One-Hot-Encode our labels
2. and count vectorize our texts

### Step 3: Checking Reliability of Language Detector

```python
text = ['si din mening', 'enig med vår anmelder?', 'lite nytt', "top marks!"]

for sentence in text:
    print(f'Lang: {detect(sentence)}  Text: {sentence}')
```

### Step 4: Preparing for Translation

The reliability of our language detector is quite poor, but it is capable of detecting english at least. Therefore, we'll settle for two functions that checks _only_ whether a text is in _english_, and if so, translates it to norwegian.

```python
translator = Translator()
unrecognized = []
norwegian = 0
failed = 0
n_translated = 0

def translate_text(text):
    global n_translated
    translated = translator.translate(text, src='en', dest='no')
    n_translated += 1
    return translated.text

def recognize_language(text):
    global norwegian, failed
    try:
        if detect(text) == 'en':
            raise Exception("Exception raised to avoid code block freezing forever, since API is down.")
            #return translate_text(text)
        else:
            norwegian += 1
            return text
    except:
        failed += 1
        unrecognized.append(text)
        return text
```

### Step 5: Preparing for Cleaning and Lemmatization

After all texts have been translated, they are then to be sent through 2 more functions that set them to lowercase, remove special symbols, and finally lemmatize the words.

```python
punctuation = '''´!"#$%&'()*+,-./:;<=>?@[\]^_`{|}~``«»-'''

nltk.download("stopwords")
stop_words = stopwords.words('norwegian')

def clean_text(text):
    text = text.lower()
    tokens = word_tokenize(text)
    tokens = [token for token in tokens if token not in stop_words and token not in punctuation]
    text = ' '.join(tokens)
    return text

nltk.download('punkt')
lemmatizer = NorwegianStemmer()

def lemmatize(text):
    text = clean_text(text)
    return lemmatizer.stem(text)
```

### Step 6: Preparing for Word/Sentiment Frequency Analysis

Finally, all text cleaned and lemmatized, a final function will assess the overall frequency of every word, and its associated sentiment.

```python
neg_words = Counter()
pos_words = Counter()
neu_words = Counter()

def count_freq(text, label):
    sentiment = label
    words = text.split()
    if sentiment == 'Negative':
        neg_words.update(words)
    elif sentiment == 'Positive':
        pos_words.update(words)
    elif sentiment == 'Neutral':
        neu_words.update(words)
```

### Step 7: Executing Preprocessing Functions

```python
def preprocess_text(data_set):
    preprocessed = []
    for row in data_set:
        text = recognize_language(row['text'])
        text = lemmatize(text)
        count_freq(text, row['label'])
        preprocessed.append(text)
    return preprocessed

text_train = preprocess_text(ch2_data['train'])
text_dev = preprocess_text(ch2_data['dev'])
text_test = preprocess_text(ch2_data['test'])
```

## Chapter 3: On Finding a Plane (Task 3 of 3)

We want to train a convolutional neural network which can detect a plane among a set of 60000 32x32 pixel color images, 50000 of which will be used to train, and the rest for development testing and final testing.

### Step 1: Loading The Dataset

```python
(train_images, train_labels), (test_images, test_labels) = keras.datasets.cifar10.load_data()
```

### Step 2: Defining Class Names and Showing Some Images

The below code was borrowed from https://www.tensorflow.org/tutorials/images/cnn

```python
class_names = ['airplane', 'automobile', 'bird', 'cat', 'deer',
               'dog', 'frog', 'horse', 'ship', 'truck']

plt.figure(figsize=(10,10))
for i in range(25):
    plt.subplot(5,5,i+1)
    plt.xticks([])
    plt.yticks([])
    plt.grid(False)
    plt.imshow(train_images[i])
    plt.xlabel(class_names[train_labels[i][0]])
plt.show()
```

### Step 3: From Multiclass To Binary Dataset

Initially, the category number for a plane is 0. However, we now change it so that a plane is represented by the number 1, while all others are represented by 0.

```python
plane_category_number = 0
train_labels = (train_labels == plane_category_number).astype(int)
test_labels = (test_labels == plane_category_number).astype(int)
```

### Step 4: Isolating Validation Data

We take 20% of our training data and use it for validation data.

```python
train_images, val_images, train_labels, val_labels = train_test_split(
    train_images, 
    train_labels, 
    test_size=0.2, 
    random_state=42)
```

### Step 5: Normalizing Pixel Values From 0-255 To 0-1

Normalizating image values should help make the training process more efficient and accurate.

```python
train_images = train_images.astype('float32') / 255.0
val_images = val_images.astype('float32') / 255.0
test_images = test_images.astype('float32') / 255.0
```

### Step 6: Training a Convolution Neural Network

We make a basic convolutional neural network of convolutional layers, max-pooling layers, a flattening layer, while ending it all with a couple of dense layers using a ReLu activation function, and finally a sigmoid activation function since it is a binary classification problem.

The code below was inspired from the website https://www.tensorflow.org/tutorials/images/cnn 

```python
model = keras.Sequential([
    layers.Conv2D(32, (3, 3), activation='relu', input_shape=(32, 32, 3)),
    layers.MaxPooling2D((2, 2)),
    layers.Conv2D(64, (3, 3), activation='relu'),
    layers.MaxPooling2D((2, 2)),
    layers.Conv2D(64, (3, 3), activation='relu'),
    layers.Flatten(),
    layers.Dense(64, activation='relu'),
    layers.Dense(1, activation='sigmoid')
])
```

```python
model.summary()
```

### Step 7: Compiling The Model

Compiling the model, we specify an optimizer, loss function, and evaluation metric to be used during training.

```python
model.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['acc'])
```

### Step 8: Training, Final Scoring and Saving

We train our model, giving it a final score with our test images, and then save it.

```python
history = model.fit(train_images, train_labels, epochs=10, validation_data=(val_images, val_labels), workers=10)
loss, acc = model.evaluate(test_images, test_labels, verbose=0)
print('Test accuracy:', acc)
model.save('my_model.h5')
```

### Step 9: Plotting Training / Validation Loss and Accuracy

10 epochs of training gave us quite the accurate model, going by our validation scores. Though even just the first epoch already gave us an accuracy of 91 and 92.6% for training and validation respectively.

We now plot the progress of our model below.

```python
plt.figure(figsize=(6, 6))
plt.plot(history.history['acc'])
plt.plot(history.history['val_acc'])
plt.title('Model accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend(['Train', 'Validation'], loc='upper left')
plt.show()

plt.figure(figsize=(6, 6))
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('Model loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Train', 'Validation'], loc='upper left')
plt.show()
```

### Step 10: Predicting /w Test and Displaying Confusion Matrix as Heatmap

Our validation scores flattened out from about epoch 4 or 5. Indicating that the model past this point might tend towards overfitting, although validation score also improved some.

Now that we're at our end, we try predicting the test images to plot a confusion matrix, and inspect the results.

```python
predictions = model.predict(test_images)
predictions = np.rint(predictions)
cm = confusion_matrix(test_labels, predictions)

plt.figure(figsize=(8, 6))
sb.heatmap(cm, annot=True, fmt='d', cmap=plt.cm.Blues, cbar=False)
plt.xlabel('Predicted label')
plt.ylabel('True label')
plt.show()
```

### Step 12: Predicting The New Images

We predict the new images with a new function. This loads our trained model, loads the new images, prepares them with resizing and normalization, and then makes our prediction, getting the likelihood of a plane, before rounding that probability, and printing the result.

```python
tf.config.experimental_run_functions_eagerly(True)

def predict_images(image_paths):
    model = keras.models.load_model('my_model.h5')
    for path in image_paths:
        image = Image.open(path)

        size = (32, 32)
        resized_image = image.resize(size)

        image_array = np.asarray(resized_image)
        
        image_array = image_array.astype('float32') / 255.0

        image_array = np.expand_dims(image_array, axis=0)

        prediction = model.predict(image_array)

        predicted_label = np.rint(prediction[0][0])

        plt.imshow(resized_image)
        plt.show()

        print(f"Predicting if plane = 1, not plane = 0 {path}: {predicted_label}")
    
predict_images(new_images)
```
