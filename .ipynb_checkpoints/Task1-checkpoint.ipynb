### Step 4: Acquiring a Method of Calculating Dependencies
The next stop for our data cleaning is to see if there are any features whose value distributions are significantly dependent on the values of another feature, which would imply redundancies.

For this step we will be using Theil's U method of calculating categorical asymmetrical dependencies, with code for the algorithm copied from this implementation example: https://www.kaggle.com/code/shakedzy/alone-in-the-woods-using-theil-s-u-for-survival

(Note: the dython python library also provides a simplified method for using Theil's U, but unfortunately we could not get this to work, and suspect that the associations() method in dython is in fact malfunctioning due to a dependency error)

The copied code includes the function conditional_entropy which is merely a stepping stone for the final function theil_u which we'll be using directly. The theil_u function takes 2 features as input, in the form of the lists x and y, and outputs the rate at which y can predict x. As such, we have an asymmetric output, requiring that we run the theil_u function over all combinations of features and _in both directions_.

```python
# copied from above-mentioned source

def conditional_entropy(x,y):
    # entropy of x given y
    y_counter = Counter(y)
    xy_counter = Counter(list(zip(x,y)))
    total_occurrences = sum(y_counter.values())
    entropy = 0
    for xy in xy_counter.keys():
        p_xy = xy_counter[xy] / total_occurrences
        p_y = y_counter[xy[1]] / total_occurrences
        entropy += p_xy * math.log(p_y/p_xy)
    return entropy

def theil_u(x,y):
    s_xy = conditional_entropy(x,y)
    x_counter = Counter(x)
    total_occurrences = sum(x_counter.values())
    p_x = list(map(lambda n: n/total_occurrences, x_counter.values()))
    s_x = ss.entropy(p_x)
    if s_x == 0:
        return 1
    else:
        return (s_x - s_xy) / s_x
```

```python
# applying Theil's U

dimensions = len(ch1_df.columns)
dependencies = np.zeros((dimensions, dimensions))

for x_index, x in enumerate(ch1_df.columns):
    for y_index, y in enumerate(ch1_df.columns):
        result = theil_u(ch1_df[x].to_list(), ch1_df[y].to_list())
        dependencies[x_index][y_index] = result
        print(f"Prediction rate of {x} using {y} is {result:.2f}.")
        
```

### Step 5: Constructing a Heatmap of Theil's U Dependency
In the last step we managed to produce some very interesting numbers, and print them. But this is a lot of numbers to deal with, so we try instead to visualize the results with a heatmap, using seaborn.

```python
heatmap_ch1_df = pd.DataFrame(dependencies)

sb.set_theme(font_scale=0.8, rc={"figure.figsize": (11,11)})
sb.heatmap(heatmap_ch1_df, annot=True, fmt=".2f", cmap="crest", xticklabels=ch1_df.columns, yticklabels=ch1_df.columns)
```

### Step 6: Analysing Theil's U Dependency
Because we have a lot of features, even the above illustration, nice as it looks, is a little much to deal with. So instead we decide to query our Theil's U dependencies for 2 key possibilities:

1. Any dependencies between features greater than 0.67
2. Any predictability of mushroom edibility greater than 0.1, by any feature.

The first query is a boundary for what we feel is an alarming level of dependency. While the second query may provide interesting findings of itself, there's another reason for us to do it. It's because what we find here may influence our decision about what to do with the results of the first query, since we would not want to take the wrong action against any feature that is already considerably contributive in predicting mushroom edibility.

```python
d_len = len(dependencies)
for x in range(d_len):
    for y in range(d_len):
        x_col = ch1_df.columns[x]
        y_col = ch1_df.columns[y]
        d_val = dependencies[x][y]
        
        # we skip looking at mushroom edibility predicting other features' values
        if(y_col != "poisonous?" and y_col != x_col):
            if(d_val >= 0.67 or (d_val >= 0.1 and d_val == "poisonous?")):
                print(f"{y_col} predicts {x_col}: {d_val}")
```

### Step 7: Evaluating Dependency Results
Of what we've found, these facts are clear:

1. "odor" is by far the most valuable predictor of whether a mushroom is edible.
2. "stalk-color-above-ring" and "stalk-color-below-ring" are massive predictors of "gill-attachment", and lesser (but still great) predictors of "veil-color"
3. "veil-color", meanwhile, is a good predictor of "gill-attachment", but interestingly, the same is also true for the reverse relationship

As such, it appears that "stalk-color-above-ring" and "stalk-color-below-ring" duplicates a lot of the information present in our dataset. The same information for which our machine learning algorithms would have as basis for building models. Meanwhile, "veil-color" and "gill-attachment" have a partial overlap, and as such there are signs of redundancy here too, while the stalk-color / veil-color / gill-attachment quadruple, together forms a local network of dependencies, such that together they show even more signs of information redundancy.

However, at this time, it is uncertain if these signs of redundancy and the value of "odor" as predictor can tell us something anything about whether any features should definitely be removed. Though, our findings are not useless, and the network of dependencies is something we can come back to later. Removing most of the quadruple could be a good experiment, for instance.

### Step 8: Machine Learning and One-Hot-Encoding
Now that we are done cleaning, we must transform our data into a more useful format. This because the machine learning algorithms we will be using, will be expecting numerical data. As such, our first step into machine learning proper, is to perform one-hot-encoding on our data. This will create new columns, in the form of pseudo-features, for every categorical value for every current feature, setting the values inside the new columns to 1 if the specific categorical value (on which basis the column was made), is present at the specified row, and to 0 otherwise.

```python
ch1_dummies= pd.get_dummies(ch1_df)
display("shape of dummies: " + str(ch1_dummies.shape))
for c in ch1_dummies.columns:
    print(c)
```

### Step 9: Isolating Data for Training and Testing
Next we must isolate our data into 3 parts:

1. A final test set, to run when all our models have been constructed. This will be 10% of our data.
2. A development data set, for testing the model during development. This will also be 10% of our data.
3. A training data set. This will be 80% of our data.

The splits will involve a randomized shuffle and stratification of the data, to negate potential patterns in how the data was recorded.

To split into 3 parts, we will have to run train_test_split 2 times. Because we want 10% each for the dev and test sets, we simply bundle these together as dev_test and give it 20% of the total set in the first split. Then, for the next split, we split the set evenly 50-50, giving us 10% of the original set size for both.

```python
ch1_X, ch1_y = ch1_dummies.loc[:, "cap-shape_b":], ch1_dummies.loc[:, "poisonous?_p"]

ch1_X_train, ch1_X_dev_test, ch1_y_train, ch1_y_dev_test = train_test_split(
    ch1_X, ch1_y,
    test_size=0.2,
    random_state=42,
    shuffle=True,
    stratify=y)

ch1_X_dev, ch1_X_test, ch1_y_dev, ch1_y_test = train_test_split(
    ch1_X_dev_test, ch1_y_dev_test,
    test_size=0.5,
    random_state=42,
    stratify=ch1_y_dev_test)
```

### Step 10: Machine Learning Plan

### Step 11: Basic RidgeClassifier

We now begin with the RidgeClassifer, as outlined in the plan.

Notably, and due to how we will be displaying the results of our training in a similar manner throughout this file, we define a function to do this for us, so that we don't have to repeat this block of code so often.

```python
ridge = RidgeClassifier()
ridge.fit(ch1_X_train, ch1_y_train)
```

```python
def show_training_results(model_title, model, X_train, y_train, X_dev, y_dev):
    model_train_pred = model.predict(X_train)
    model_dev_pred = model.predict(X_dev)

    print(f"{model_title} - training set")
    print(f"Accuracy: {accuracy_score(y_train, model_train_pred):.3}")
    print(f"Confusion matrix:\n{confusion_matrix(y_train, model_train_pred)}")

    print("\n")

    print(f"{model_title} - dev set")
    print(f"Accuracy: {accuracy_score(y_dev, model_dev_pred):.3}")
    print(f"Confusion matrix:\n{confusion_matrix(y_dev, model_dev_pred)}")
```

```python
show_training_results("Ridge", ridge, ch1_X_train, ch1_y_train, ch1_X_dev, ch1_y_dev)
```

### Step 12: Grid Search and Cross-validation over RidgeClassifier

A simple default training already provides results at or adjacent to 100%. In fact, if we hadn't set the precision display value at 3 or higher decimals, both would round up to 100%. Meanwhile, the confusion matrix shows that we only mislabelled a single datapoint in both cases. With such success, it is difficult to imagine we could improve on it. Still, we'll try to run a grid search for curiosity's sake. And some cross-validation to see if we might've been lucky, despite randomizing and stratifying our data sets.

Because we'll be doing grid search with cross-validation repeatedly over the different algorithms, we'll also make a couple of functions here to both train and display the GridSearchCV.

```python
def train_gscv(estimator, params):
    gscv = GridSearchCV(
        estimator=estimator,
        param_grid=params,
        scoring="accuracy",
        n_jobs=-1,
        cv=10,
        return_train_score=True)
    gscv.fit(ch1_X_train, ch1_y_train)
    return gscv
```

```python
def show_gscv_results(model_title, gscv):
    print(f"Best params: {gscv.best_params_}")
    print(f"Best score: {gscv.best_score_:.3f}")
    mean_training_score = gscv.cv_results_['mean_train_score']
    print(f"Mean Training Score: {np.around(mean_training_score, decimals=3)}")
    mean_testing_score = gscv.cv_results_['mean_test_score']
    print(f"Mean Testing Score: {np.around(mean_testing_score, decimals=3)}")
    
    print("\n")

    show_training_results(
        model_title,
        gscv.best_estimator_,
        ch1_X_train, ch1_y_train,
        ch1_X_dev, ch1_y_dev)
```

```python
params = {
    "alpha": [0, 0.001, 0.01, 0.1, 1.0, 10.0],
    "solver": ["svd", "cholesky", "sparse_cg", "lsqr"],
    "random_state": [42]
}

ridge_gscv = train_gscv(RidgeClassifier(), params)
```

```python
show_gscv_results("Ridge/Grid Search", ridge_gscv)
```

### Step 13: Observations from Grid Search /w Ridge

We can derive 2 interesting finds from our grid search, plus we should add 1 disclaimer.

Starting with the disclaimer: the amount of data that our grid search had to use for training was significantly less here, than the data we used to train our first default ridge model with. Here we also used the training data, BUT, the cross-validation algorithm always took away 10% for internal testing. Knowing this though, we can use this fact to make our first interesting observation:

1. A small but not insignificant change in the amount of training data makes no difference to the result.
2. As for the second interesting find: the best alpha value appears to be 0, in which case the algorithm should work pretty much like ordinary least squares, as there's no L2 regularization going on.

As for the cross-validation mean score arrays, the only thing of note here is that in a few instances a very small change occurs in the amount of mistakes. A difference of 0.1%, which in the end is quite negligible, but points towards parameter choices making at least some difference, which could scale over a larger dataset.

### Step 22: Inspecting The Decision Trees

Before we run a grid search, it would be interesting to get an idea of the sort of decision trees we have already produced with our perfect fit random forest.

For, what may the properties of these trees tell us? In particular, how deep do they go? How many leaves do they have? We can calculate a few such basic statistics with the code below.

```python
num_estimators = 0

sum_leaves = 0
max_leaves = 0

sum_depth = 0
max_depth = 0

for estimator in rfc.estimators_:
    num_estimators += 1
    t = estimator.tree_
    sum_depth += t.max_depth
    if(max_depth < t.max_depth):
        max_depth = t.max_depth
    sum_leaves += t.n_leaves
    if(max_leaves < t.n_leaves):
        max_leaves = t.n_leaves
        
avg_depth = sum_depth / num_estimators
avg_leaves = sum_leaves / num_estimators

print(f"Num Estimators: {num_estimators}")
print(f"Max Depth: {max_depth}")
print(f"Avg Depth: {avg_depth}")
print(f"Max Leaves: {max_leaves}")
print(f"Avg Leaves: {avg_leaves}")
```

A tenfold or higher increase in the number of samples, and some guarantees that the samples recorded are relevant to the ecosystem where we ourselves collect our own samples for prediction, could well sway the argument in favor of a "yes" though.
