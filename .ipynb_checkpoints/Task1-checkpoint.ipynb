```python
pd_series1 = df.loc[:, "veil-type"]
pd_series1.hist()
```

```python
p_counter = 0
not_p_counter = 0
for item in pd_series1:
    if (item == "p"):
        p_counter += 1
    else:
        not_p_counter += 1
print("num of p: " + str(p_counter))
print("num of not-p: " + str(not_p_counter))
```
